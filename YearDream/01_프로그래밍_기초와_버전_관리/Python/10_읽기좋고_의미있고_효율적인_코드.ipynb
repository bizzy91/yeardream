### 줄바꿈

적절한 줄바꿈은 코드의 가독성에 큰 영향을 미칩니다. 예를 들어 아래처럼 코드가 따닥따닥 붙어있으면 코드가 한눈에 들어오지 않습니다.

```Python
class Bread(): 
    taste = "밀가루"
    def eat(self): 
        print(self.taste, "맛이 나요!")
redBean = Bread() 
choux = Bread() 
redBean.taste = "팥"
choux.taste = "슈크림"
redBean.eat()
choux.eat()
```

그렇다고 오른쪽 코드처럼 필요 없는 줄바꿈을 많이 하면 쓸데없이 코드 라인수만 길어집니다.

파이썬에서는 한 두번의 줄바꿈을 통해 연관된 개념을 묶어서 나타내길 권장하고 있습니다.

* 모듈 가져오기(import) 아래, 최상위 레벨 메서드, 클래스 정의부는 두 번의 줄바꿈을 이용해 구분합니다.
* 클래스 안의 메소드는 한 번의 줄바꿈을 이용해 구분합니다.
* 서로 연관된 함수와 코드 덩어리는 여러 번의 줄바꿈을 통해 구분하기도 하며 함수 내에서 논리적인 문단을 구분하기 위해 쓰이기도 합니다.

줄바꿈은 새로운 개념을 시작한다는 새로운 단서입니다. 줄바꿈을 통해 연관된 코드를 구분하면 묶여있는 코드가 하는 역할을 한눈에 파악할 수 있습니다.

```Python
if greeting:
```

## 캡슐화

객체지향 프로그래밍이란 프로그램 명령어들을 하나의 독립된 단위인 객체로 보고 코딩을 하는 것입니다.

앞에서 실습한 다형성도 객체지향 프로그래밍의 요소 중 하나이며 지금 배우는 캡슐화와 상속도 객체지향 프로그래밍의 요소입니다.

캡슐화는 가지고 있는 데이터를 바깥에서 은닉하고 일부 함수만 바깥에 노출함으로써 데이터 접근을 관리하는 것입니다.

구체적으로 설명하자면 클래스 내 존재하는 멤버 변수를 클래스 내 메서드를 통해서만 설정하고 반환할 수 있게 하는 것이 캡슐화입니다.

---

### 지시사항
1. Human 클래스의 멤버 변수 __name과 __birthday를 매개변수로 받아 값을 수정하는 __set() 함수를 만드세요.
2. Human 클래스의 멤버 변수 __name을 반환하는 get_name() 함수를 만드세요.
3. Human 클래스의 멤버 변수 __birthday를 반환하는 get_birthday() 함수를 만드세요.

Tips
* 파이썬에서 메서드 두 개의 밑줄 문자를 통해 외부 접근을 보호합니다. 즉, 선언한 __set() 함수와 두 멤버 변수는 클래스 밖에서 호출하면 안 됩니다.
* 다른 언어에서는 private, public과 같은 접근 지정자로 외부의 접근을 막을 수 있습니다.

````python
class Human:
    def __init__(self, name, birthday):
        self.__set(name, birthday)

    def __set(self, name, birthday):
        self.__name = name
        self.__birthday = birthday

    def get_name(self):
        return self.__name
    
    def get_birthday(self):
        return self.__birthday
    
H = Human('bizzy', 'Nov. 25th')
print(H.get_name)
print(H.get_birthday)
````

## 상속

앞선 다형성 실습에서 해보았듯이 상속이란 부모 클래스의 멤버를 자식 클래스에 물려주는 것을 말합니다.

클래스의 재사용을 위해서 상속을 사용해야 하지만 상속은 원하지 않는 메서드까지 상속되는 문제가 있습니다.

상속 안티 패턴을 해결하기 위한 방법으로 파이썬의 컴포지션(Composition)이 있습니다.

````python
class FindTen:
    def __init__(self, number):
        self.number = number

    def two(self):
        return self.number % 2

    def five(self):
        return self.number % 5


class FindSix:
    def __init__(self, number):
        self.number = number
        self.find_five = FindTen(number)

    def two(self):
        return self.number % 2

    def three(self):
        return self.number % 3

    def five(self):
        return self.find_five.five()
    
FS = FindSix(10)
print(FS.two())
print(FS.three())
print(FS.five())
````


````python
import math

def print_area(radius):
    area = math.pi * radius * radius
    print("원의 넓이:", area)

def get_area(radius):
    area = math.pi * radius * radius
    return area
    
def get_circle_area(radius):
    area = math.pi * radius * radius
    print("원의 넓이:", area)
    return area

result = get_area(3)
print_area(3)
if result > 20:
    print("더 큰 원입니다.")
elif result > 10:
    print("큰 원입니다.")
else:
    print("작은 원입니다.")
````
