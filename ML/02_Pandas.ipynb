### List, Dictionary, Numpy

```python
import numpy as np
```

```python
# 리스트/배열로 DataFrame 만들기
col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
df_list1 = pd.DataFrame(list1, columns=col_name1)
df_array1 = pd.DataFrame(array1, columns=col_name1)

print('DataFrame <- List\n', df_list1)
print('DataFrame <- Array\n', df_array1)
```

```python
col_name2 = ['col1', 'col2', 'col3']
list2 = [
    [1, 2, 3],
    [4, 5, 6]    
]
array2 = np.array(list2)
df_list2 = pd.DataFrame(list2, columns=col_name2)
df_array2 = pd.DataFrame(array2, columns=col_name2)

print('DataFrame <- List\n', df_list2)
print('DataFrame <- Array\n', df_array2)
```

```python
dictionary = {
    'col1': [1, 4],
    'col2': [2, 5],
    'col3': [3, 6]
}
df_dict = pd.DataFrame(dictionary)

print('DataFrame <- Dictionary\n', df_dict)
```

```python
array3 = df_dict.values
print('DataFrame -> Array\n', array3)
```

```python
list3 = df_dict.values.tolist()
print('DataFrame -> List\n', list3)
```

```python
dictionary2 = df_dict.to_dict()
print('DataFrame -> Dictionary\n', dictionary2)
print('--------------------to_dict("series")--------------------\n', df_dict.to_dict('series'))
print('--------------------to_dict("split")--------------------\n', df_dict.to_dict('split'))
print('--------------------to_dict("records")--------------------\n', df_dict.to_dict('records'))
print('--------------------to_dict("index")--------------------\n', df_dict.to_dict('index'))
```

### 결손 처리

```python
# DataFrame.fillna('XXX'), 결손 데이터 'XXX'로 채우기
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df
```

```python
# 'Age' 컬럼의 결손 값을 'Age'의 평균으로 채운다.
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df
```

### Apply lambda

```python
def get_square(a):
    return a*a

print('The square of 3:', get_square(3))

lambda_square = lambda x: x**2
print('(lambda) The square of 3:', lambda_square(3))
```

```python
a = [1, 2, 3]
squares = map(lambda_square, a)
list(squares)
```

```python
# 'Name' 칼럼의 값에 'Name' 값의 길이를 입력한다.
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df.head()
```

```python
# 'Child_Adult'컬럼의 값에 15세 미만은 'Child', 그렇지 않으면 'Adult'를 입력한다.
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head()
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else: cat = 'Elderly'
        
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
